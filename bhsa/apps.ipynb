# Apps

A Text-Fabric *app* is a Python module that contain specific functionality for a given corpus.

```python
%load_ext autoreload
%autoreload 2
```

```python
from tf.app import use
```

```python
A = use('uruk:clone', hoist=globals())
```

```python
A = use('bhsa:clone', hoist=globals())
```

```python
results = A.search('''
clause
  phrase function=Pred
    word pdp#verb
''')
```

```python
A.table(results, end=4)
```

```python
A = use('oldbabylonian:clone', hoist=globals())
```

```python
results = A.search('''
line
  =: sign reading=um
  <: sign reading=ma
  <: word
     := sign reading=ma
''')
```

```python
A.table(results, end=4, fmt='layout-orig-unicode')
```
